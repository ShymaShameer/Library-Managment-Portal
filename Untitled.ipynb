```python
from pptx import Presentation
from pptx.util import Inches
from PIL import Image, ImageDraw, ImageFont

import os

out_dir = "/mnt/data/azure_podcast_ppt"
os.makedirs(out_dir, exist_ok=True)

def make_slide_image(title, shape_type, filename):
    img = Image.new("RGB", (1200, 675), "white")
    draw = ImageDraw.Draw(img)

    # Title
    draw.text((40, 30), title, fill="black")

    # Simple icons
    if shape_type == "storage":
        draw.rectangle([200, 150, 1000, 550], outline="blue", width=8)
        draw.text((500, 350), "Storage", fill="blue")
    elif shape_type == "blob":
        draw.ellipse([250, 150, 950, 550], outline="purple", width=10)
        draw.text((520, 350), "Blob", fill="purple")
    elif shape_type == "lake":
        draw.polygon([(200,500),(600,200),(1000,500)], outline="teal", width=10)
        draw.text((520, 350), "OneLake", fill="teal")
    elif shape_type == "files":
        for i in range(3):
            draw.rectangle([300+i*200, 200, 450+i*200, 500], outline="green", width=6)
        draw.text((520, 520), "Files", fill="green")
    elif shape_type == "tables":
        for i in range(4):
            draw.line([(250,200+i*100),(950,200+i*100)], fill="orange", width=4)
            draw.line([(250+i*175,200),(250+i*175,600)], fill="orange", width=4)
        draw.text((520, 620), "Tables", fill="orange")
    elif shape_type == "cosmos":
        draw.ellipse([250,150,950,550], outline="red", width=6)
        draw.line([(250,350),(950,350)], fill="red", width=4)
        draw.line([(600,150),(600,550)], fill="red", width=4)
        draw.text((520, 350), "Cosmos DB", fill="red")

    path = os.path.join(out_dir, filename)
    img.save(path)
    return path

slides_info = [
    ("Azure Storage", "storage", "storage.png"),
    ("Azure Blob Storage", "blob", "blob.png"),
    ("OneLake", "lake", "lake.png"),
    ("Azure Files", "files", "files.png"),
    ("Azure Tables", "tables", "tables.png"),
    ("Azure Cosmos DB", "cosmos", "cosmos.png"),
]

image_paths = []
for t, s, f in slides_info:
    image_paths.append(make_slide_image(t, s, f))

prs = Presentation()

for img_path, (title, _, _) in zip(image_paths, slides_info):
    slide = prs.slides.add_slide(prs.slide_layouts[6])
    slide.shapes.add_picture(img_path, Inches(1), Inches(1), width=Inches(8))

output_pptx = "/mnt/data/Azure_Storage_Podcast_Presentation.pptx"
prs.save(output_pptx)

output_pptx
```
