### Enhancing SQL Control with engine.connect()

#### Using SQLAlchemy, for all four CRUD operations

#### CREATE (Inserting data)

```python
## Inserting a new friend to the friends table

# Creating an engine to connect to the database
engine = create_engine(connection_string)

# Defining the insert data
insert_data = [
    {
        "first_name": "James", 
        "last_name": "Taylor", 
        "email": "james@email.com", 
        "phone": "5551006", 
        "max_loan": 2, 
        "notes": "Frequent borrower"
    }
]

# Executing the insert  
with engine.begin() as connection:
    connection.execute(
        text("""
            INSERT INTO friends
            (first_name, last_name, email, phone, max_loan, notes)
            VALUES
            (:first_name, :last_name, :email, :phone, :max_loan, :notes)
        """), 
        insert_data
    )

     
```

#### READ (Select)

```python
## Viewing all the friends 

# Executing the select query

with engine.connect() as connection:
    result= connection.execute(
        text("SELECT * FROM friends")
    )

    for row in result:
        print(row)
```

#### UPDATE

```python
book_alchemist = books_from_sql[books_from_sql['title'].str.contains('alchemist', case=False)]
book_alchemist
```

```python
## Updating the genre of book_id 2

# Creating an engine to connect to the database
engine = create_engine(connection_string)

# Defining the update query
update_query = """
    UPDATE books
    SET genre = "Fantasy"
    WHERE book_id = 2;
"""

# Executing the update query with an explicit commit
with engine.connect() as connection:
    transaction = connection.begin()
    try:
        connection.execute(text(update_query))
        transaction.commit()
    except:
        transaction.rollback()
        raise
```

```python
## Updating the book as unavailable

with engine.begin() as connection:
    connection.execute(
        text("""
            UPDATE books
            SET is_available = FALSE
            WHERE book_id = :book_id
        """),
        {"book_id": 5}
    )
```

#### DELETE

```python
## Deleting the loan record for book_id 9

with engine.begin() as connection:
    connection.execute(
        text("""
            DELETE FROM loans
            WHERE book_id = :book_id
        """),
        {"book_id": 9}
    )
```
